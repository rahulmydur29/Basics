```python
import requests   #request library reads the url and loads the webpage
from bs4 import BeautifulSoup as bs   #importing the beautiful soup library
```

```python
url="https://www.amazon.in/Samsung-Galaxy-Ultra-Burgundy-Storage/product-reviews/B09SZT1K1K/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
```

```python
page=requests.get(url)  #extract the url and load the data data of webpage
page

#Response 200 is all good
#Byt response 503 is error ....re run the cell multiple times to fix it.provided url must be valid
```

```python
page.content  #prints the source code of the webpage
```

```python
soup=bs(page.content,'html.parser')  #prints the source code in beautiful manner.using beautiful soup
```

```python
soup 
```

```python
print(soup.prettify())    #indentation everything will be as it is
```

```python
names=soup.find_all('span',class_="a-profile-name")[2:]
names
```

```python
len(names)
```

```python
import pandas as pd
df=pd.DataFrame(names,columns={"Customer Name"})
df
```

```python
df=df.drop_duplicates(ignore_index=True)  
#if ignore_index=true is not mentioned then along with customer names index will alse be removed for that particular customer it becomes like0,2,4 etc index gets deleted.
df
```

```python
r_title=soup.find_all('a',class_="review-title")
r_title
```

```python
len(r_title)
```

```python
r_title[0].get_text()
```

```python
r_title[0].get_text()[1:-1]
```

```python
review_title=[]
for i in range(0,len(r_title)):
  review_title.append(r_title[i].get_text()[1:-1])
review_title
```

```python
dates=soup.find_all('span',class_="review-date")[2:]
dates
#dates=soup.select('span.review-date')  alternate method to find all
```

```python
len(dates)
```

```python
dates[0].get_text()
```

```python
dates[0].get_text().replace("Reviewed in India on",'')
```

```python
review_date=[]
for i in range(0,len(dates)):
  review_date.append(dates[i].get_text().replace("Reviewed in India on",''))
review_date
```

```python
rating=soup.select('i.review-rating')[2:]
rating
```

```python
len(rating)
```

```python
rating[0].get_text()
```

```python
review_rating=[]
for i in range(0,len(dates)):
  review_rating.append(rating[i].get_text().replace("out of 5 stars",''))
review_rating
```

```python
content=soup.select('span.review-text')
content
```

```python
content[0].get_text()[4:-3]
```

```python
review_text=[]
for i in range(0,len(content)):
  review_text.append(content[i].get_text()[4:-3])
review_text
```

```python
df['review_title']=review_title
df['review_date']=review_date
df['review_rating']=review_rating
df['review_text']=review_text
df
```
