```python
df=df.drop(columns=['Bedroom2','Postcode','Lattitude','Longtitude'])
df.shape
```

```python
df.isnull().sum()
```

```python
df['Landsize']=df['Landsize'].fillna(df['Landsize'].mean())
df['BuildingArea']=df['BuildingArea'].fillna(df['BuildingArea'].mean())
```

```python
df.isnull().sum()
```

```python
df['Bathroom']=df['Bathroom'].fillna(df['Bathroom'].mode().values[0])  #since mode is in series .values[0] is necessary
df['Car']=df['Car'].fillna(df['Car'].mode().values[0])
df['Distance']=df['Distance'].fillna(df['Distance'].mode().values[0])
df['Propertycount']=df['Propertycount'].fillna(df['Propertycount'].mode().values[0])
```

```python
df.isnull().sum()
```

```python
df[['YearBuilt']]  #first find the age of the building by subtracting present year and yearbuilt.then fill nan values with mode of age of the building
```

```python
df['YearBuilt']=2021-df['YearBuilt']
```

```python
df['YearBuilt']=df['YearBuilt'].fillna(df['YearBuilt'].mode().values[0])
```

```python
df.isnull().sum()
```

```python
df['Price']=df['Price'].fillna(df['Price'].mean())
df.isnull().sum()
```

```python
df=df.dropna()  #drops the rows containing NaN values
df.isnull().sum()
```

```python
df2=df.copy().select_dtypes(exclude='object')
df2
```

```python
x=df2.drop('Price',axis=1).values
y=df2['Price'].values
```
