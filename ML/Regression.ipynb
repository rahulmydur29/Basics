# Linear Regression using Formula Method:

```python
import numpy as np
from matplotlib import pyplot as plt

# Sample data
x = np.array([2, 4, 6, 8, 10])
y = np.array([3, 5, 6, 8, 9])

# Using the formula method
x_mean = np.mean(x)
y_mean = np.mean(y)
x_deviation = x - x_mean
y_deviation = y - y_mean
m_formula = np.sum(x_deviation * y_deviation) / np.sum(x_deviation**2)
b_formula = y_mean - m_formula * x_mean
y_predicted_formula = m_formula * x + b_formula

# Plotting
plt.scatter(x, y, color='blue', label='Original Data')
plt.plot(x, y_predicted_formula, color='red', label='Regression Line (Formula)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression using Formula Method')
plt.legend()
plt.grid(axis="x")
plt.show()

# Print the results
print("Results using Formula Method:")
print(f"Slope (m): {m_formula}")
print(f"Intercept (b): {b_formula}")
```

# Linear Regression using scikit-learn:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
x = np.array([2, 4, 6, 8, 10])
y = np.array([3, 5, 6, 8, 9])

# Using scikit-learn
x_reshaped = x.reshape(-1, 1)
reg = LinearRegression()
reg.fit(x_reshaped, y)
m_sklearn = reg.coef_[0]
b_sklearn = reg.intercept_
y_predicted_sklearn = reg.predict(x_reshaped)

# Plotting
plt.scatter(x, y, color='blue', label='Original Data')
plt.plot(x, y_predicted_sklearn, color='green', label='Regression Line (scikit-learn)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression using scikit-learn')
plt.legend()
plt.grid(True)
plt.show()

# Print the results
print("Results using scikit-learn:")
print(f"Slope (m): {m_sklearn}")
print(f"Intercept (b): {b_sklearn}")
```

# Bad Fit:

```python
# These values for the x- and y-axis should result in a very bad fit for linear regression:

import matplotlib.pyplot as plt
from scipy import stats

x = [89,43,36,36,95,10,66,34,38,20,26,29,48,64,6,5,36,66,72,40]
y = [21,46,3,35,67,95,53,72,58,10,26,34,90,33,38,20,56,2,47,15]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

# Multiple Regression

```python
import numpy as np

# Sample data
x1 = np.array([2, 4, 6, 8, 10])
x2 = np.array([3, 5, 7, 9, 11])
y = np.array([6, 8, 10, 12, 14])

# Using the formula method
X = np.vstack((np.ones_like(x1), x1, x2)).T
coeff_matrix = np.linalg.inv(X.T @ X) @ X.T @ y

# Coefficients for each predictor variable
intercept = coeff_matrix[0]
coeff_x1 = coeff_matrix[1]
coeff_x2 = coeff_matrix[2]

# Print the results
print("Results using Formula Method:")
print(f"Intercept (b0): {intercept}")
print(f"Coefficient for x1 (m1): {coeff_x1}")
print(f"Coefficient for x2 (m2): {coeff_x2}")
```

```python
# Multiple Linear Regression using scikit-learn with Pearson Correlation:
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr

# Sample data
x1 = np.array([2, 4, 6, 8, 10])
x2 = np.array([3, 5, 7, 9, 11])
y = np.array([6, 8, 10, 12, 14])

# Using scikit-learn
X = np.vstack((x1, x2)).T
reg = LinearRegression()
reg.fit(X, y)
coefficients_sklearn = reg.coef_
y_predicted_sklearn = reg.predict(X)

# Calculate Pearson correlation for each feature
correlation_x1, _ = pearsonr(x1, y)
correlation_x2, _ = pearsonr(x2, y)

# Plotting
plt.scatter(y, y_predicted_sklearn, color='blue')
plt.plot([min(y), max(y)], [min(y), max(y)], color='red', linestyle='--')
plt.xlabel('Actual Y')
plt.ylabel('Predicted Y (scikit-learn)')
plt.title('Multiple Linear Regression using scikit-learn')
plt.grid(True)
plt.show()

# Print the results
print("Results using scikit-learn:")
print(f"Coefficients (m1, m2): {coefficients_sklearn}")
print(f"Pearson Correlation (x1): {correlation_x1}")
print(f"Pearson Correlation (x2): {correlation_x2}")
```
