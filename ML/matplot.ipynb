```python
#imorting matplotlib module
import matplotlib.pyplot as plt
# from matplotlib import pyplot as plt
```

```python
#x=axis values
x=[1,3,5,7,9]  # list no: of values in x and y must be same or else error will be generated
```

```python
#y=axis values
y=[10,9,5,8,4]  #list
```

```python
#line graph from list
plt.plot(x,y)  # plot() is used to draw line graph
```

```python
#line graph from tuples
a=(1,2,3,4,5,6)
b=(9,4,5,7,2,4)
plt.plot(a,b)   # a is x axis and b is y axis,(a,b) as (x,y) is drawn on the graph example (1,9) (2,4)
```

```python
plt.plot(a,b)
plt.show()  #avoids the line printing above the graph in case of only plt.plot()
```

```python
import numpy as np
arr1=np.array([1,2,3,4,5])
arr2=np.array([6,7,8,9,10])
plt.plot(arr1,arr2)
plt.show()
```

```python
# for bar graph
arr1=np.array([1,2,3,4,5])
arr2=np.array([4,2,8,6,9])
plt.bar(arr1,arr2)
plt.show()
```

```python
arr1=np.array([1,2,3,4,5])
arr2=np.array([4,2,8,6,9])
plt.barh(arr1,arr2)
plt.show()
```

```python
#scatter graph
arr1=np.array([1,2,3,4,5])
arr2=np.array([4,2,8,6,9])
plt.scatter(arr1,arr2)
plt.show()
```

```python
a1=[1,3,5,7,9]
a2=(4,6,8,10,12)
a3=np.array([1,9,15,18,20]) #it is not nesecessarily to be in same data structure,all three can be either one of them list or tuple or array
```

```python
plt.plot(a1,a2)
plt.plot(a1,a3)
plt.plot(a2,a3) #3 lines in graph representing each plot in the same graph
plt.show()
```

```python
plt.figure(figsize=(6,6))  # figsize is the parameter that takes scale of x and y axis to increase the scale of the graph both horizontal and vertical 
plt.plot(a1,a2)
plt.plot(a1,a3)
plt.plot(a2,a3) 
plt.show()
```

```python
plt.figure()  # figure() is the method to increase the scale of the graph both horizontal and vertical and takes figsize as the parameter
plt.plot(a1,a2)
plt.plot(a1,a3)
plt.plot(a2,a3) 
plt.show()
```

```python
age=[18,22,24,26,28]
height=[150,154,158,145,160]
```

```python
plt.plot(age,height)
plt.show()
plt.bar(age,height)
plt.show()
plt.barh(age,height)
plt.show()
plt.scatter(age,height)
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.plot(age,height)
plt.show()
plt.bar(age,height)
plt.show()
plt.barh(age,height)
plt.show()
plt.scatter(age,height)
plt.show()

```

```python
#to customize the graph by namig x and y axis and title
plt.figure(figsize=(6,4))
plt.plot(age,height)
plt.xlabel("Average age of people")
plt.ylabel("Average height of people")
plt.title("Age vs Height")
plt.show()
```

```python
bangalore=[100.6,158.61,305.54,394.96,724.79]  #dependent
mysore=[10.5,25.21,58.65,119.27,274.87]
year=[1972,1982,1992,2001,2012]    #year independent
```

```python
plt.figure(figsize=(7,5))
plt.plot(year,bangalore,c="r",label="Bangalore")  # independent on x axis and dependent variable on y axis
plt.plot(year,mysore,c="k",label="Mysore")
plt.xlabel("Years",fontsize=13)
plt.ylabel("Electricity consumption",fontsize=13)
plt.title("Electricity consumption trend analysis over the years",fontsize=14)
plt.legend() # to enable label lines in the graph
plt.plot()
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.plot(year,bangalore,c="r",label="Bangalore")  # independent on x axis and dependent variable on y axis
plt.plot(year,mysore,c="k",label="Mysore")
plt.xlabel("Years",fontsize=13)
plt.ylabel("Electricity consumption",fontsize=13)
plt.title("Electricity consumption trend analysis over the years",fontsize=14)
plt.legend(loc="upper right") # to enable label lines in the graph
plt.plot()
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.plot(year,bangalore,c="r",label="Bangalore",marker="*")  # marker to highlight the ordered pairs
plt.plot(year,mysore,c="k",label="Mysore",marker=".")
plt.xlabel("Years",fontsize=13)
plt.ylabel("Electricity consumption",fontsize=13)
plt.title("Electricity consumption trend analysis over the years",fontsize=14)
plt.legend(loc="upper right") # to enable label lines in the graph
plt.plot()
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.plot(year,bangalore,c="r",label="Bangalore",marker="*",linestyle="dashed") #marker to highlight the ordered pairs
plt.plot(year,mysore,c="k",label="Mysore",marker=".",linestyle="dashdot")
plt.xlabel("Years",fontsize=13)
plt.ylabel("Electricity consumption",fontsize=13)
plt.title("Electricity consumption trend analysis over the years",fontsize=14)
plt.legend(loc="upper right") # to enable label lines in the graph
plt.plot()
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.plot(year,bangalore,c="r",label="Bangalore",marker="*")  # marker to highlight the ordered pairs
plt.plot(year,mysore,c="k",label="Mysore",marker=".")
plt.xlabel("Years",fontsize=13)
plt.ylabel("Electricity consumption",fontsize=13)
plt.title("Electricity consumption trend analysis over the years",fontsize=14)
plt.legend(loc="upper right") # to enable label lines in the graph
plt.xticks(year)  #to change x axis scale and look
plt.plot()
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.plot(year,bangalore,c="r",label="Bangalore",marker="*")  # marker to highlight the ordered pairs
plt.plot(year,mysore,c="k",label="Mysore",marker=".")
plt.xlabel("Years",fontsize=13)
plt.ylabel("Electricity consumption",fontsize=13)
plt.title("Electricity consumption trend analysis over the years",fontsize=14)
plt.legend(loc="upper right") # to enable label lines in the graph
#plt.xticks(year,rotation=45)
plt.xticks(np.arange(1970,2014,4),rotation=45)
plt.plot()
plt.show()
```

```python
cars=['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data=[23,17,35,29,12,41]
```

```python
cars=['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data=[23,17,35,29,12,41]
color=['Orange','Red','Green','Blue','Cyan','Magenta']
explode=[0,0,0,0,0,0]
```

```python
plt.figure(figsize=(7,6))
plt.pie(data,labels=cars,colors=color,counterclock=False,startangle=90,shadow=True,autopct="%1.1f%%",explode=explode)
plt.show()
```

```python
cars=['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data=[23,17,35,29,12,41]
color=['Orange','Red','Green','Blue','Cyan','Magenta']
explode=[0.2,0,0,0,0.4,0]
plt.figure(figsize=(7,6))
plt.pie(data,labels=cars,colors=color,counterclock=False,startangle=90,shadow=True,autopct="%1.1f%%",explode=explode)
plt.show()
```

```python
cars=['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data=[23,17,35,29,12,41]
color=['Orange','Red','Green','Blue','Cyan','Magenta']
explode=[0.2,0,0,0,0.4,0]
plt.figure(figsize=(7,6))
plt.pie(data,labels=cars,colors=color,counterclock=True,startangle=90,shadow=True,autopct="%1.1f%%",explode=explode,wedgeprops={"edgecolor":"black","linewidth":2})
plt.show()
```

```python
plt.figure(figsize=(7,6))
plt.pie(data,labels=cars,colors=color,counterclock=False,startangle=90,shadow=False,autopct="%1.1f%%",explode=explode,wedgeprops={"edgecolor":"black","linewidth":2})
plt.legend()
plt.show()
```
