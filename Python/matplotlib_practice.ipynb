# Bar Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["python","C","C++","Java"])
y = np.array([85,50,75,83])
plt.xlabel("language", fontsize=15)
plt.ylabel("Number", fontsize=15)
plt.title("Bar Chart", fontsize=15)
c=["r","b","y","m"]
plt.bar(x,y, width=0.2,color=c,align="center",edgecolor="r", label="popularity" )#(linewidth=5, linestyle=":",alpha=0.4)
plt.legend()
plt.show()
```

```python
#horizontal bar plot
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["python","C","C++","Java"])
y = np.array([85,50,75,83])
plt.xlabel("language", fontsize=15)
plt.ylabel("Number", fontsize=15)
plt.title("Bar Chart", fontsize=15)
c=["r","b","y","m"]
plt.barh(x,y,color=c,align="center",edgecolor="r" ) #(linewidth=5, linestyle=":",alpha=0.4)
plt.show()
```

# scatter plot

```python
import matplotlib.pyplot as plt
import numpy as np

day = np.array([1,2,3,4,5,6,7])
no = np.array([2,3,4,7,3,2,3])

plt.title("scatter plot", fontsize=15)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)

c=["r","g","b","y","m","k","c"]
s=[50,500,70,80,600,100,300]

plt.scatter(day,no,color=c,s=s,alpha=0.5) #alpha=tranperency
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

day = np.array([1,2,3,4,5,6,7])
no = np.array([2,3,4,7,3,2,3])

plt.title("scatter plot", fontsize=15)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)

c=["r","g","b","y","m","k","c"]
s=[50,500,70,80,600,100,300]

plt.scatter(day,no,color=c,s=s, marker="*")
# markers
# .	point
# , 	pixel
# o	circle
# v	triangle_down
# ^	triangle_up
# <	triangle_left
# >	triangle_right
# 1	tri_down
# 2	tri_up
# 3	tri_left
# 4	tri_right
plt.show()
```

```python
#pie chart
x2=[10,20,30,40]
y2=["c","c++","java","python"]
ex=[0.3,0.0,0.0,0.0]
c=["r","g","y","m"]

plt.pie(x2,labels=y2,explode=ex, colors=c, autopct="%0.1f%%", shadow=True, radius=1,
        startangle=90, textprops={"fontsize":12}, counterclock=False, wedgeprops={"linewidth":5},
       rotatelabels=True)
plt.title("Pie Chart")
plt.legend(loc=1)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Labels for different sections of the pie chart
labels = ['Apples', 'Bananas', 'Oranges', 'Grapes']

# Sizes (proportions) of each section
sizes = [35, 25, 20, 20]

# Colors for each section
colors = ['red', 'yellow', 'orange', 'purple']

# Explode (pull out) a section
explode = (0.1, 0, 0, 0)

# Create a pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode=explode)

# Equal aspect ratio ensures that the pie is drawn as a circle.
plt.axis('equal')

# Add a title
plt.title("Fruit Distribution")

# Show the plot
plt.show()
```
