```python
marks=np.random.randint(20,50,size=(4,4))  # or marks=np.random.randint(20,50,(4,4))
marks
```

```python
Subjects=['Math','Computers','Biology','Sports']
names=['Peter','Josh','Rob','John']
```

```python
df=pd.DataFrame(marks,index=names,columns=Subjects)
df
```

```python
df['Math']
```

```python
df[['Math']]
```

```python
#to add columns
df['Total']=df['Math']+df['Computers']+df['Biology']+df['Sports']
df
```

```python
df['percentage']=(df['Total']/200)*100
df
```

```python
type (df)
```

```python
#to drop columns
df1=df.drop('percentage',axis=1)# axis=1 represents columns(removal of columns) and axis=0 is x axis
df1
```

```python
df2=df1.drop("Josh",axis=0)   #assigning to a variable and then print is compulsory
df2
```

```python
d={'Cricket':[1,3,np.nan,4,6,7,2,np.nan],
   'Baseball':[5,np.nan,np.nan,5,7,4,7,3],
   'Tennis':[1,2,3,4,5,6,7,8]}
```

```python
df=pd.DataFrame(d)
df
```

```python
#handling of missing values
#drop the rows or columns containing missing values  (dropna)
# 1 fill the missing values with 0  (fillna)
# 2 fill the missing values with mean or median or mode (fillna)
```

```python
df1=df.dropna(axis=0)
df1
```

```python
df2=df.dropna(axis=1)
df2
```

```python
df.fillna(value=0)
```

```python
df['Cricket']=df['Cricket'].fillna(value=df['Cricket'].median())
df
```

```python
df['Baseball']=df['Baseball'].fillna(value=df['Baseball'].median())
df
```

```python
df=df.set_index([['alpha','bravo','charlie','delta','echo','foxtrot','golf','hotel']])
df
#df.reset_index(inplace=True)
```
