```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/drive/MyDrive/aiml/Mall_Customers.csv")
df
```

```python
x=df.iloc[:,3:5].values
x
```

```python
#below block of code is to determine the best value of k(no: of clusters)
#we can determine k using elbow method
#max value of k should not exceed the square root of total numbers of rows
```

```python
k=range(2,15)
from sklearn.cluster import KMeans
SSE=[]
for i in k:
  model_demo=KMeans(n_clusters=i)
  model_demo.fit(x)
  sse=model_demo.inertia_ 
  SSE.append(sse)

plt.plot(k,SSE)
plt.show()
```

```python
#we will assume no:of clusters(k) as 5 because elbow is steepest at 5
```

```python
k=5
from sklearn.cluster import KMeans
model=KMeans(n_clusters=k,random_state=0)
model.fit(x)
```

```python
y=model.predict(x)
y
```
